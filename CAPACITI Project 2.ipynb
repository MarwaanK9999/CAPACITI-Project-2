```python
import matplotlib.pyplot as plt 

data = pd.read_csv('owid-covid-data.csv')
data = data[(data["date"] == "2022-07-18")]

# x axis values 
x = data['total_deaths']
# corresponding y axis values 
y = data['location'] 

plt.figure(figsize=(25, 45))

# plotting the points  
plt.plot(x, y) 
  
# naming the x axis 
plt.xlabel('Total Number of Deaths') 
# naming the y axis 
plt.ylabel('Country') 
 
    
    
# giving a title to my graph 
plt.title('Total Deaths Around The World') 
  
# function to show the plot 
plt.show()
```

```python
import matplotlib.pyplot as plt 

data = pd.read_csv('owid-covid-data.csv')
data = data[(data["date"] == "2022-07-18")]

data1 = data[(data["location"] == "High income")]
data2 = data[(data["location"] == "Upper middle income")]
data3 = data[(data["location"] == "Lower middle income")]

frames = [data1, data2, data3]

result = pd.concat(frames)

df = pd.DataFrame(result, columns=["location", "total_cases", "total_deaths", "new_deaths"])
df.plot(x="location", y=["total_cases", "total_deaths", "new_deaths"], kind="bar", title="Comparison of Data in Certain Income Classes in Countries", figsize=(9, 8))
```

```python
import matplotlib.pyplot as plt 

data = pd.read_csv('owid-covid-data.csv')
data = data[(data["date"] == "2022-07-18")]

data1 = data[(data["location"] == "Europe")]
data2 = data[(data["location"] == "Asia")]
data3 = data[(data["location"] == "Africa")]
data4 = data[(data["location"] == "North America")]
data5 = data[(data["location"] == "South America")]
data6 = data[(data["location"] == "Oceana")]
data7 = data[(data["location"] == "World")]

frames = [data1, data2, data3, data4, data5, data6, data7]

result = pd.concat(frames)

df = pd.DataFrame(result, columns=["location", "total_cases", "total_deaths", "new_deaths"])
df.plot(x="location", y=["total_cases", "total_deaths", "new_deaths"], title="Comparison of Data in Different Continents", kind="bar", figsize=(9, 8))
```

```python
import matplotlib.pyplot as plt 

data = pd.read_csv('owid-covid-data.csv')
dataBeginning = data[(data["date"] == "2020-02-24")]
dataLatest = data[(data["date"] == "2022-07-18")]

data1 = dataBeginning[(dataBeginning["location"] == "World")]
data2 = dataLatest[(dataLatest["location"] == "World")]

frames = [data1, data2]

result = pd.concat(frames)

df = pd.DataFrame(result, columns=["location", "total_cases", "total_deaths", "new_deaths"])
df.plot(x="location", y=["total_cases", "total_deaths", "new_deaths"], kind="bar", title="Comparison of Data Beginning and Current Figures", figsize=(9, 8))

df = pd.DataFrame(result, columns=["location", "total_cases", "total_deaths", "new_deaths"])
df.plot(x="location", y=["total_cases", "total_deaths", "new_deaths"], kind="hist", title="Comparison of Data Beginning and Current Figures", figsize=(9, 8))
```
